For better analysis, I delete orders that do not have an assigned user id or category.

```python
df.dropna(subset=['user_id','category_code'],inplace=True)
df
```

```python
df[df.columns[df.isna().any()]].isna().sum()
```

```python
df = df.reset_index(drop=True)
df
```

```python
df['event_time']=pd.to_datetime(df['event_time'])
df.info()
df
```

```python
temp_df = df['category_code'].str.rsplit('.',n=1, expand=True)
df = df.drop(columns='category_code')
df['product_category'] = temp_df[1]
df
```

```python
df.groupby(df['event_time'].dt.year).count()
```

```python
df[df['event_time'].dt.year==1970].sort_values('event_time')
```

```python
df = df[df['event_time'].dt.year>1970].sort_values('event_time').reset_index(drop=True)
df
```
